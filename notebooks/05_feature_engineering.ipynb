```python
# ==================================================
# Import Libraries
# ==================================================

import pandas as pd
import numpy as np
import os
```

```python
# ==================================================
# Connect Google Drive
# ==================================================

from google.colab import drive
drive.mount('/content/drive')
```

```python
# ==================================================
# Project Paths
# ==================================================

PROJECT_PATH = "/content/drive/MyDrive/student-productivity-analytics"

CLEANED_DATA_PATH = f"{PROJECT_PATH}/data/cleaned"

PROCESSED_DATA_PATH = f"{PROJECT_PATH}/data/processed"

os.makedirs(PROCESSED_DATA_PATH, exist_ok=True)

print("Paths Ready!")
```

```python
# ==================================================
# Load Cleaned Dataset
# ==================================================

dataset_path = (
    f"{CLEANED_DATA_PATH}/"
    "student_productivity_cleaned.csv"
)

df = pd.read_csv(dataset_path)

print("Dataset Loaded Successfully!")
```

```python
# ==================================================
# Preview Dataset
# ==================================================

df.head()
```

```python
# ==================================================
# Productivity Efficiency Score
# ==================================================

df["Productivity_Efficiency"] = (
    df["Productivity_Score"] /
    (df["Study_Hours"] + 1)
)

df["Productivity_Efficiency"] = (
    df["Productivity_Efficiency"]
    .round(2)
)

print("Productivity Efficiency Created!")
```

```python
# ==================================================
# Digital Distraction Score
# ==================================================

df["Digital_Distraction_Score"] = (
    df["Screen_Time_Hours"] * 0.6
    -
    df["Study_Hours"] * 0.3
    +
    (10 - df["Focus_Score"]) * 0.4
)

df["Digital_Distraction_Score"] = (
    df["Digital_Distraction_Score"]
    .round(2)
)

print("Distraction Score Created!")
```

```python
# ==================================================
# Wellness Index
# ==================================================

df["Wellness_Index"] = (
    df["Sleep_Hours"] * 0.4
    +
    df["Mood_Score"] * 0.3
    -
    df["Stress_Level"] * 0.3
)

df["Wellness_Index"] = (
    df["Wellness_Index"]
    .round(2)
)

print("Wellness Index Created!")
```

```python
# ==================================================
# Study Consistency Score
# ==================================================

study_std = (
    df.groupby("Student_ID")["Study_Hours"]
    .transform("std")
)

df["Study_Consistency_Score"] = (
    10 - study_std
)

df["Study_Consistency_Score"] = (
    df["Study_Consistency_Score"]
    .clip(0, 10)
    .round(2)
)

print("Consistency Score Created!")
```

```python
# ==================================================
# Sleep Deficit Score
# ==================================================

df["Sleep_Deficit"] = (
    8 - df["Sleep_Hours"]
)

df["Sleep_Deficit"] = (
    df["Sleep_Deficit"]
    .clip(lower=0)
    .round(2)
)

print("Sleep Deficit Created!")
```

```python
# ==================================================
# Burnout Risk Score
# ==================================================

df["Burnout_Risk_Score"] = (
    df["Stress_Level"] * 0.4
    +
    df["Sleep_Deficit"] * 0.3
    +
    df["Screen_Time_Hours"] * 0.2
    +
    (10 - df["Mood_Score"]) * 0.1
)

df["Burnout_Risk_Score"] = (
    df["Burnout_Risk_Score"]
    .round(2)
)

print("Burnout Risk Score Created!")
```

```python
# ==================================================
# Performance Category
# ==================================================

conditions = [
    df["Exam_Score"] >= 80,
    df["Exam_Score"] >= 60,
    df["Exam_Score"] >= 40
]

categories = [
    "Excellent",
    "Good",
    "Average"
]

df["Performance_Category"] = np.select(
    conditions,
    categories,
    default="Poor"
)

print("Performance Categories Created!")
```

```python
# ==================================================
# High Risk Flag
# ==================================================

df["High_Risk_Flag"] = np.where(
    (
        (df["Burnout_Risk_Score"] > 7)
        &
        (df["Exam_Score"] < 50)
    ),
    1,
    0
)

print("High Risk Flag Added!")
```

```python
# ==================================================
# New Feature Preview
# ==================================================

df[[
    "Productivity_Efficiency",
    "Digital_Distraction_Score",
    "Wellness_Index",
    "Study_Consistency_Score",
    "Burnout_Risk_Score",
    "Performance_Category"
]].head()
```

```python
# ==================================================
# Save Processed Dataset
# ==================================================

processed_path = (
    f"{PROCESSED_DATA_PATH}/"
    "student_productivity_processed.csv"
)

df.to_csv(processed_path, index=False)

print("Processed Dataset Saved!")
```
