```python
import pandas as pd
import numpy as np
import random

from faker import Faker
from datetime import datetime, timedelta
```

```python
# =========================================================
# RANDOM SEED FOR REPRODUCIBILITY
# =========================================================

SEED = 42

random.seed(SEED)

np.random.seed(SEED)

Faker.seed(SEED)
```

```python
fake = Faker()
```

```python
# ==================================================
# Dataset Parameters
# ==================================================

NUM_STUDENTS = 500
NUM_DAYS = 30

START_DATE = datetime(2025, 1, 1)

print(f"Students: {NUM_STUDENTS}")
print(f"Tracking Days: {NUM_DAYS}")
```

```python
# ==================================================
# Student Personas
# ==================================================

personas = [
    "High Performer",
    "Burnout",
    "Distracted",
    "Average",
    "Inconsistent"
]

print(personas)
```

```python
# ==================================================
# Empty Dataset List
# ==================================================

data = []
```

```python
# ==================================================
# Synthetic Data Generation
# ==================================================

for student_id in range(1, NUM_STUDENTS + 1):

    persona = random.choice(personas)

    age = random.randint(18, 25)

    gender = random.choice(["Male", "Female"])

    department = random.choice([
        "Computer Science",
        "Information Technology",
        "Electronics",
        "Mechanical",
        "Business"
    ])

    year = random.randint(1, 4)

    for day in range(NUM_DAYS):

        current_date = START_DATE + timedelta(days=day)

        # ------------------------------------------
        # Persona-based behavior generation
        # ------------------------------------------

        if persona == "High Performer":

            sleep = np.random.normal(7.5, 0.8)
            study = np.random.normal(5.5, 1)
            screen = np.random.normal(4, 1)
            stress = np.random.normal(4, 1)
            focus = np.random.normal(8, 1)
            mood = np.random.normal(8, 1)

        elif persona == "Burnout":

            sleep = np.random.normal(4.5, 1)
            study = np.random.normal(8, 1.5)
            screen = np.random.normal(7, 1)
            stress = np.random.normal(9, 1)
            focus = np.random.normal(5, 1)
            mood = np.random.normal(4, 1)

        elif persona == "Distracted":

            sleep = np.random.normal(6, 1)
            study = np.random.normal(2.5, 1)
            screen = np.random.normal(10, 2)
            stress = np.random.normal(6, 1)
            focus = np.random.normal(3, 1)
            mood = np.random.normal(6, 1)

        elif persona == "Average":

            sleep = np.random.normal(6.5, 1)
            study = np.random.normal(4, 1)
            screen = np.random.normal(6, 1.5)
            stress = np.random.normal(5, 1)
            focus = np.random.normal(6, 1)
            mood = np.random.normal(6, 1)

        else:

            sleep = np.random.normal(5.5, 2)
            study = np.random.normal(4, 3)
            screen = np.random.normal(7, 3)
            stress = np.random.normal(6, 2)
            focus = np.random.normal(5, 2)
            mood = np.random.normal(5, 2)

        # ------------------------------------------
        # Clip values to realistic ranges
        # ------------------------------------------

        sleep = np.clip(sleep, 2, 10)
        study = np.clip(study, 0, 14)
        screen = np.clip(screen, 0, 16)
        stress = np.clip(stress, 1, 10)
        focus = np.clip(focus, 1, 10)
        mood = np.clip(mood, 1, 10)

        # ------------------------------------------
        # Derived Metrics
        # ------------------------------------------

        exam_score = (
            study * 5
            + focus * 4
            + sleep * 2
            - stress * 3
            + np.random.normal(0, 5)
        )

        exam_score = np.clip(exam_score, 0, 100)

        attendance = (
            sleep * 10
            + focus * 5
            - stress * 2
            + np.random.normal(0, 5)
        )

        attendance = np.clip(attendance, 40, 100)

        assignment_completion = (
            focus * 10
            + study * 5
            + np.random.normal(0, 10)
        )

        assignment_completion = np.clip(
            assignment_completion,
            0,
            100
        )

        productivity_score = (
            focus * 4
            + study * 3
            + mood * 2
            - stress * 2
        )

        productivity_score = np.clip(
            productivity_score,
            0,
            100
        )

        # ------------------------------------------
        # Burnout Logic
        # ------------------------------------------

        if stress > 8 and sleep < 5:
            burnout = "High"

        elif stress > 6:
            burnout = "Medium"

        else:
            burnout = "Low"

        # ------------------------------------------
        # Append Row
        # ------------------------------------------

        data.append({

            "Student_ID": student_id,

            "Date": current_date,

            "Persona": persona,

            "Age": age,

            "Gender": gender,

            "Department": department,

            "Year_of_Study": year,

            "Sleep_Hours": round(sleep, 1),

            "Study_Hours": round(study, 1),

            "Screen_Time_Hours": round(screen, 1),

            "Stress_Level": round(stress, 1),

            "Focus_Score": round(focus, 1),

            "Mood_Score": round(mood, 1),

            "Attendance_Percentage": round(attendance, 1),

            "Assignment_Completion": round(
                assignment_completion,
                1
            ),

            "Productivity_Score": round(
                productivity_score,
                1
            ),

            "Exam_Score": round(exam_score, 1),

            "Burnout_Risk": burnout
        })
```

```python
# ==================================================
# Create DataFrame
# ==================================================

df = pd.DataFrame(data)

print("Dataset Created Successfully!")
```
