```python
# ==================================================
# Import Libraries
# ==================================================

import pandas as pd
import numpy as np
import os
```

```python
# ==================================================
# Connect Google Drive
# ==================================================

from google.colab import drive
drive.mount('/content/drive')
```

```python
# ==================================================
# Project Paths
# ==================================================

PROJECT_PATH = "/content/drive/MyDrive/student-productivity-analytics"

PROCESSED_DATA_PATH = f"{PROJECT_PATH}/data/processed"

DASHBOARD_PATH = f"{PROJECT_PATH}/dashboard"

os.makedirs(DASHBOARD_PATH, exist_ok=True)
```

```python
# ==================================================
# Load ML Ready Dataset
# ==================================================

dataset_path = (
    f"{PROCESSED_DATA_PATH}/"
    "student_productivity_ml_ready.csv"
)

df = pd.read_csv(dataset_path)

print("Dataset Loaded!")
```

```python
# ==================================================
# Dataset Preview
# ==================================================

df.head()
```

```python
# ==================================================
# Dashboard Audience
# ==================================================

dashboard_audience = [
    "University Administration",
    "Academic Advisors",
    "Student Wellness Teams",
    "Students",
    "Researchers"
]

dashboard_audience
```

```python
# ==================================================
# Executive KPIs
# ==================================================

avg_exam_score = round(
    df["Exam_Score"].mean(),
    2
)

avg_productivity = round(
    df["Productivity_Score"].mean(),
    2
)

avg_burnout = round(
    df["Burnout_Risk_Score"].mean(),
    2
)

high_risk_students = len(
    df[df["High_Risk_Flag"] == 1]
)

print("Average Exam Score:", avg_exam_score)

print("Average Productivity:", avg_productivity)

print("Average Burnout Risk:", avg_burnout)

print("High Risk Students:", high_risk_students)
```

```python
# ==================================================
# Export Dashboard Dataset
# ==================================================

dashboard_dataset_path = (
    f"{DASHBOARD_PATH}/"
    "dashboard_dataset.csv"
)

df.to_csv(
    dashboard_dataset_path,
    index=False
)

print("Dashboard Dataset Exported!")
```
