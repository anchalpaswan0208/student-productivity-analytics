```python
# ==================================================
# Import Libraries
# ==================================================

import pandas as pd
import numpy as np
import os
```

```python
# ==================================================
# Connect Google Drive
# ==================================================

from google.colab import drive
drive.mount('/content/drive')
```

```python
# ==================================================
# Project Paths
# ==================================================

PROJECT_PATH = "/content/drive/MyDrive/student-productivity-analytics"

RAW_DATA_PATH = f"{PROJECT_PATH}/data/raw"

CLEANED_DATA_PATH = f"{PROJECT_PATH}/data/cleaned"

os.makedirs(CLEANED_DATA_PATH, exist_ok=True)

print("Paths Ready!")
```

```python
# ==================================================
# Load Raw Dataset
# ==================================================

dataset_path = f"{RAW_DATA_PATH}/student_productivity_dataset.csv"

df = pd.read_csv(dataset_path)

print("Dataset Loaded Successfully!")
```

```python
# ==================================================
# Preview Dataset
# ==================================================

df.head()
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
# ==================================================
# Introduce Missing Values
# ==================================================

for col in [
    "Sleep_Hours",
    "Study_Hours",
    "Mood_Score"
]:
    df.loc[
        df.sample(frac=0.02).index,
        col
    ] = np.nan

print("Missing Values Introduced!")
```

```python
# ==================================================
# Missing Values After Injection
# ==================================================

df.isnull().sum()
```

```python
# ==================================================
# Handle Missing Values
# ==================================================

df["Sleep_Hours"].fillna(
    df["Sleep_Hours"].median(),
    inplace=True
)

df["Study_Hours"].fillna(
    df["Study_Hours"].median(),
    inplace=True
)

df["Mood_Score"].fillna(
    df["Mood_Score"].median(),
    inplace=True
)

print("Missing Values Handled!")
```

```python
# ==================================================
# Duplicate Check
# ==================================================

df.duplicated().sum()
```

```python
# ==================================================
# Duplicate Check
# ==================================================

df.duplicated().sum()
```

```python
# ==================================================
# Outlier Inspection
# ==================================================

df[
    (df["Study_Hours"] > 12)
]
```

```python
# ==================================================
# Feature Engineering
# ==================================================

df["Study_Efficiency"] = (
    df["Exam_Score"] /
    (df["Study_Hours"] + 1)
)

print("Feature Created!")
```

```python
# ==================================================
# Wellness Score
# ==================================================

df["Wellness_Score"] = (
    df["Sleep_Hours"] * 0.4
    + df["Mood_Score"] * 0.3
    - df["Stress_Level"] * 0.3
)

print("Wellness Score Added!")
```

```python
# ==================================================
# Date Conversion
# ==================================================

df["Date"] = pd.to_datetime(df["Date"])

print("Date Converted!")
```

```python
# ==================================================
# Final Validation
# ==================================================

df.info()
```

```python
# ==================================================
# Final Dataset Preview
# ==================================================

df.head()
```

```python
# ==================================================
# Save Cleaned Dataset
# ==================================================

cleaned_path = (
    f"{CLEANED_DATA_PATH}/"
    "student_productivity_cleaned.csv"
)

df.to_csv(cleaned_path, index=False)

print("Cleaned Dataset Saved!")
```
